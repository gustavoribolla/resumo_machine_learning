# Support Vector Machines (SVM)

SVM é um modelo que busca encontrar o hiperplano que melhor separa as classes, maximizando a margem entre elas. A margem é a distância entre o hiperplano e o ponto mais próximo de cada classe. O hiperplano é a linha que separa as classes. 

O SVM é um modelo de aprendizado supervisionado, que pode ser usado tanto para classificação quanto para regressão. Neste notebook, vamos focar na classificação.

O SVM é um modelo poderoso, mas que pode ser sensível a hiperparâmetros. O hiperparâmetro mais importante é o `C`, que controla a penalidade por classificações erradas. Um `C` alto significa que o modelo vai tentar classificar todos os pontos corretamente, mesmo que isso signifique uma margem menor. Um `C` baixo significa que o modelo vai tentar maximizar a margem, mesmo que isso signifique classificar alguns pontos errados.

- Hard SVM (C alto): tenta classificar todos os pontos corretamente, mesmo que isso signifique uma margem menor. Pode levar a overfitting. Hard Margin SVM assume que os dados são linearmente separáveis sem erros. Busca maximizar a margem sem permitir classificações incorretas.

- Soft SVM (C baixo): tenta maximizar a margem, mesmo que isso signifique classificar alguns pontos errados. Pode levar a underfitting. Soft Margin SVM permite classificações incorretas, mas tenta minimizá-las. Soft Margin SVM permite alguns erros de classificação para lidar com dados não perfeitamente separáveis. Introduz o parâmetro de penalização C que controla o trade-off entre maximizar a margem e minimizar o erro.



```python
from sklearn.svm import SVC

# Dados de treino
X_train = [[0, 0], [1, 1]]
y_train = [0, 1]


# Hard Margin SVM (não especificando C, usa valor alto por padrão)
modelo_hard = SVC(kernel='linear', C=1e10)
modelo_hard.fit(X_train, y_train)

# Soft Margin SVM
modelo_soft = SVC(kernel='linear', C=1.0)
modelo_soft.fit(X_train, y_train)
```
